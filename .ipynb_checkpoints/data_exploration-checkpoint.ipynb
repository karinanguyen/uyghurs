```python
sources['month'] = pd.to_datetime(sources['month'], format="%Y-%d-%m") 
sources['year']= sources['month'].dt.year
```

```python
country_overtime = sources.drop(columns=["id","display_name", "url", "month"])
country_overtime
```

```python
#country_overtime = country_overtime.groupby(["year", "country"]).sum()
#country_overtime
```

```python
#country_overtime.plot(figsize=(20,40))
fig, ax = plt.subplots()
ax.set_xticks(country_overtime.year.unique())
country_overtime.groupby(["year", "country"]).sum()['uyghur_count'].unstack().plot(ax=ax)
```
